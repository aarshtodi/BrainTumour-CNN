```python
import numpy as np
from matplotlib import pyplot as plt
import glob
from PIL import Image
import cv2
from numpy import asarray 
```

```python
image_list_no = []
enhanced_image_no = []
image_list_yes = []
enhanced_image_yes = []
```

```python
for filename in glob.glob('/Users/Aarsh/Desktop/AScratch/Augmented/no/*.jpg'):
    print(filename)
    img = cv2.imread(filename,0)
    image_list_no.append(img)
```

```python
for filename in glob.glob('/Users/Aarsh/Desktop/AScratch/Augmented/yes/*.jpg'):
    print(filename)
    img = cv2.imread(filename,0)
    image_list_yes.append(img)
```

```python
def Enhancement(img):
    m,n = img.shape
    h = []
    size = m*n
    p = float(1.0)/size
    for i in range(256):
        h.append(0)
    nk =[]
    graylevel =[]
    for i in range(m):
        for j in range(n):
            r =img[i][j]
            if r not in graylevel:
                graylevel.append(r)
    graylevel.sort()
    L = len(graylevel)
    for i  in range(len(graylevel)):
        nk.append(0)
    for i in range(m):
        for j in range(n):
            r = img[i][j]
            pos = graylevel.index(r)
            nk[pos] = 1+ nk[pos]
    for i in range(len(nk)):
        value = graylevel[i]
        count = nk[i]
    pdf =[]
    size =m*n
    for i in range(L):
        value = (float(nk[i])/size)
        pdf.append(value)
    sum=0
    cdf = []
    for i in range(len(pdf)):
        sum =sum+ pdf[i]
        cdf.append(sum)
    newcdf = []
    for i in cdf:
        newcdf.append(int(i*(L-1)))
    for i in range(len(pdf)):
        value = graylevel[i]
        count = newcdf[i]
    newImg = np.zeros(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            try:
                newImg[i][j] = newcdf[img[i][j]]
            except:
                continue
    return newImg
```

```python
for image in image_list_no:
    #image = image.convert('LA')
    #ima = asarray(image)
    #print(ima.shape)
    img = Enhancement(image)
    enhanced_image_no.append(img)
```

```python
for image in image_list_yes:
    img = Enhancement(image)
    enhanced_image_yes.append(img)
```
