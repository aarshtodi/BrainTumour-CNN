```python
import cv2
import numpy as np
import glob
from PIL import Image
```

```python
def Open(img):
    kernel = np.ones((5,5),np.uint8)
    opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    return opening
```

```python
image_list_no = []
morpho_no = []
image_list_yes = []
morpho_yes = []
```

```python
for filename in glob.glob('/Users/Aarsh/Desktop/AScratch/Segmented/no/*.jpg'):
    print(filename)
    img = cv2.imread(filename,0)
    image_list_no.append(img)
```

```python
for image in image_list_no:
    img = Open(image)
    morpho_no.append(img)
```

```python
for i in range(len(morpho_no)):
    new = Image.fromarray(morpho_no[i])
    new = new.convert('RGB')
    new.save('{}{}{}'.format('/Users/Aarsh/Desktop/AScratch/Morphology/no/N',i+1,'.jpg'))
```
